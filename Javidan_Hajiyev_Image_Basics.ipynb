#### TASK: Open the *dog_backpack.jpg* image from the DATA folder and display it in the notebook. Make sure to correct for the RGB order.

```python

#import all necessary libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read the image
image_path = "dog_backpack.png"
image = cv2.imread(image_path)
plt.imshow(image)
plt.axis("on")  # Hide axis
plt.show()
```

```python
#correct the color distribution
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
def equalize_histogram(img):
    img_yuv = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    img_yuv[:,:,0] = cv2.equalizeHist(img_yuv[:,:,0])
    return cv2.cvtColor(img_yuv, cv2.COLOR_YUV2RGB)
image_eq = equalize_histogram(image_rgb)
plt.imshow(image_eq)
plt.axis("on")
plt.show()
```

### plot the different channels: red, green, blue

```python
# Display Red channel
# Split the image into Red, Green, and Blue channels
r, g, b = cv2.split(image_rgb)

# Create a copy of the image to manipulate the channels
image_red_only = np.zeros_like(image_rgb)
image_red_only[:, :, 0] = r

# Display
plt.figure(figsize=(8, 8))
plt.imshow(image_red_only)
plt.title("Red Channel")
plt.show()
```

```python
# Display Green channel
# Split the image into Red, Green, and Blue channels
r, g, b = cv2.split(image_rgb)

# Create a copy of the image to manipulate the channels
image_green_only = np.zeros_like(image_rgb)
image_green_only[:, :, 1] = r

# Display
plt.figure(figsize=(8, 8))
plt.imshow(image_green_only)
plt.title("Green Channel")
plt.show()
```

```python
# Display Red channel
# Split the image into Red, Green, and Blue channels
r, g, b = cv2.split(image_rgb)

# Create a copy of the image to manipulate the channels
image_blue_only = np.zeros_like(image_rgb)
image_blue_only[:, :, 2] = r

# Display
plt.figure(figsize=(8, 8))
plt.imshow(image_blue_only)
plt.title("Blue Channel")
plt.show()
```

### Resize the image to involves the half number of pixels of the original one

```python
# Resize
# Get the original image dimensions
height, width = image_rgb.shape[:2]
# Calculate new dimensions (half the original)
new_width = int(width / 2)
new_height = int(height / 2)
# Resize the image
resized_image = cv2.resize(image_rgb, (new_width, new_height))
```

```python
#plot
plt.imshow(resized_image)
plt.title("Resized Image (Half Pixels)")
plt.show()
```

#### TASK: Flip the image upside down and display it in the notebook.

```python
# Flip the image upside down
flipped_image = cv2.flip(image_rgb, 0)

# Display the flipped image
plt.imshow(flipped_image)
plt.title("Flipped Image")
plt.show()
```

#### TASK: Draw an empty RED rectangle around the dogs face and display the image in the notebook.

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Define the coordinates of the rectangle (top-left and bottom-right corners)
x1, y1 = 200, 725  # Adjust these values to fit your image
x2, y2 = 600, 375 # Adjust these values to fit your image

# Draw the rectangle on the image
cv2.rectangle(image_rgb, (x1, y1), (x2, y2), (255, 0, 0), 4)  # (255, 0, 0) is for red color, 2 is thickness

# Display the image with the rectangle
plt.imshow(image_rgb)
plt.title("Image with Rectangle")
plt.show()
```

#### TASK: Draw a BLUE TRIANGLE in the middle of the image. The size and angle is up to you, but it should be a triangle (three sides) in any orientation.

```python
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
# Define triangle vertices
pt1 = (600,375)
pt2 = (200,375)
pt3 = (420,750)

# Create triangle points array
triangle_cnt = np.array( [pt1, pt2, pt3] )

cv2.drawContours(image_rgb, [triangle_cnt], 0, (0, 0, 255), 5) #border thickness 5


# Display the image
plt.imshow(image_rgb)
plt.title("Image with Blue Triangle")
plt.show()
```
