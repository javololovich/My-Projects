## Image Thresholding

```python
# import libraries: cv2, numpy, pyplot
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read rainbow image
image_path = "rainbow.jpg"
image = np.array(cv2.imread(image_path))
```

```python
# Plot img
plt.imshow(image)
plt.axis("on")
plt.show()
```

```python
# Adding the 0 flag to read it in black and white
image_gray = cv2.imread('rainbow.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
# plot img
plt.imshow(image_gray, cmap='gray')
plt.show()
```

## Different Threshold Types

https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html

### Binary

```python
# Use threshold function to convert gray values to black or white depending on thresholds, param: THRESH_BINARY
ret, thresh1 = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY)
```

```python
ret
```

```python
# plot thresh1,cmap='gray'
plt.imshow(thresh1, cmap='gray')
plt.show()
```

### Binary Inverse

```python
# Use threshold function to convert gray values to black or white depending on thresholds, param: THRESH_BINARY_INV
ret, thresh2 = cv2.threshold(image_gray, 127, 255, cv2.THRESH_BINARY_INV)

# plot thresh2, cmap='gray'
plt.imshow(thresh2, cmap='gray')
plt.show()
```

###  Threshold Truncation

```python
# Use threshold function to convert gray values to black or white depending on thresholds, param: THRESH_TRUNC
ret, thresh3 = cv2.threshold(image_gray, 127, 255, cv2.THRESH_TRUNC)

# plot thresh3, cmap='gray'
plt.imshow(thresh3, cmap='gray')
plt.show()
```

### Threshold to Zero

```python
# Use threshold function to convert gray values to black or white depending on thresholds, param: THRESH_TOZERO
ret, thresh4 = cv2.threshold(image_gray, 127, 255, cv2.THRESH_TOZERO)

# plot thresh4, cmap='gray'
plt.imshow(thresh4, cmap='gray')
plt.show()
```

### Threshold to Zero (Inverse)

```python
# Use threshold function to convert gray values to black or white depending on thresholds, param: THRESH_TOZERO_INV
ret, thresh5 = cv2.threshold(image_gray, 127, 255, cv2.THRESH_TOZERO_INV)

# plot thresh5, cmap='gray'
plt.imshow(thresh5, cmap='gray')
plt.show()
```

# Real World Applications

## Adaptive Thresholding



### Sudoku Image

```python
# Read crossword image
img = cv2.imread('crossword.jpg',0)
```

```python
# simple function to plot an image img with a pre-defined figure size
def show_pic(img):
    fig = plt.figure(figsize=(15,15))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap='gray')
```

```python
show_pic(img)
```

## Simple Binary

```python
# Use threshold function to convert gray values to black or white depending on thresholds, param: THRESH_BINARY
ret, th1 = cv2.threshold(img, 127, 255, cv2.THRESH_BINARY)
```

```python
show_pic(th1)
```

### Adaptive Threshold

https://stackoverflow.com/questions/28763419/adaptive-threshold-parameters-confusion

    @param src Source 8-bit single-channel image.
    .   @param dst Destination image of the same size and the same type as src.
    .   @param maxValue Non-zero value assigned to the pixels for which the condition is satisfied
    .   @param adaptiveMethod Adaptive thresholding algorithm to use, see #AdaptiveThresholdTypes.
    .   The #BORDER_REPLICATE | #BORDER_ISOLATED is used to process boundaries.
    .   @param thresholdType Thresholding type that must be either #THRESH_BINARY or #THRESH_BINARY_INV,
    .   see #ThresholdTypes.
    .   @param blockSize Size of a pixel neighborhood that is used to calculate a threshold value for the
    .   pixel: 3, 5, 7, and so on.
    .   @param C Constant subtracted from the mean or weighted mean (see the details below). Normally, it
    .   is positive but may be zero or negative as well.

```python
# Use adaptiveThreshold function to convert gray values to black or white image, param: THRESH_BINARY
# Use these values: 11, and 8 for input variables
th2 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_MEAN_C, cv2.THRESH_BINARY, 11, 8)
show_pic(th2)
```

```python
# Use adaptiveThreshold function to convert gray values to black or white image, param: THRESH_BINARY
# Use these values: 15, and 8 for input variables
# cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY
th3 = cv2.adaptiveThreshold(img, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C, cv2.THRESH_BINARY, 15, 8)
```

```python
show_pic(th3)
```

```python
# Use addWeighted to blend th1 and th2 images with alpha=0.7, beta=0.3, gamma=0
blended = cv2.addWeighted(th1, 0.7, th2, 0.3, 0)
show_pic(blended)
```
